```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from matplotlib.pyplot import plot
```

```python
df = pd.read_csv('data/archive/sign_mnist/sign_mnist.csv')
```

```python
df
```

```python
label_unique = df.label.unique()
label_unique.sort()
print(label_unique)
```

```python
print(df.label.value_counts()/df.label.sum())
expected = df.label.value_counts().sum()/label_unique.shape
print(f'Математическое ожидание: {expected[0]}')
disp = (df.label.value_counts().max() - df.label.value_counts().min())/df.label.value_counts().max()
print(f'Расстояние в %: {disp}')
```

```python
X_data, X_valid, y_data, y_valid = train_test_split(df.drop(['label'], axis=1), df['label'], test_size=0.1, random_state=78)
```

```python
valid = pd.concat([y_valid, X_valid], axis=1)
dataset = pd.concat([y_data, X_data], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['label'], axis=1), df['label'], test_size=0.2, random_state=78)
```

```python
train = pd.concat([y_train, X_train], axis=1)
test = pd.concat([y_test, X_test], axis=1)
```
